```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
data = pd.read_csv('/content/Titanic_Data.csv')
```

```python
data.isnull().sum()
```

```python
data = data.drop(columns ='Cabin' , axis = 1)
```

```python
data.isnull().sum()
```

```python
data['Age'].fillna(data['Age'].mean() , inplace = True)
data.isnull().sum()
```

```python
data['Embarked'].fillna(data['Embarked'].mode()[0] , inplace = True)
data.isnull().sum()
```

```python
data.head()
```

```python
data['Embarked'].value_counts()
```

```python
data['Sex'].value_counts()
```

```python
data.replace({'Embarked':{"S" : 1 , "C" : 1 , "Q" : 2} , "Sex" : {"male" : 1 , "female" : 0}} , inplace = True)
```

```python
data
```

```python
X = data.drop(columns = ['PassengerId' , 'Survived' , 'Name' , 'Ticket'] , axis = 1)
Y = data['Survived']
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(X,Y,test_size = 0.2,stratify=Y , random_state = 1)
```

```python
model = LogisticRegression()
model.fit(X_train , Y_train)
```

```python
X_predict = model.predict(X_train)
training_accuracy = accuracy_score(X_predict , Y_train)
print(training_accuracy)
X_predict_test = model.predict(X_test)
test_accuracy = accuracy_score(X_predict_test, Y_test)
print(test_accuracy)
```

```python
input = (3 , 	1	, 22.000000	, 1 ,	0	,7.2500	, 1)
inputarray = np.asarray(input)
reshaped_array = inputarray.reshape(1,-1)
prediction = model.predict(reshaped_array)
print(prediction)

if prediction[0] == 1 :
  print("Survived")
else :
  print("Dead")
```
