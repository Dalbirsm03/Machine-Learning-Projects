This data is too unbalanced so we create a new data and then we train the model

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_card_data = pd.read_csv("/content/creditcard.csv")
credit_card_data
```

```python
credit_card_data.describe()
```

```python
credit_card_data.isnull().sum()
```

```python
credit_card_data["Class"].value_counts()
```

```python
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

As number of legit is more than fruad we will create a sample dataset of legit with same number as fraud

```python
legit_sample = legit.sample(n = 437)
```

```python
print(legit_sample)
```

```python
final_data = pd.concat([legit_sample , fraud] , axis = 0)
print(final_data)
```

Now we will be splitting the data for X and Y

```python
X = final_data.drop(columns = ["Class"] , axis = 1)
Y = final_data["Class"]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
model = LogisticRegression()
model.fit(X_train , Y_train)
```

Training Accuracy

```python
X_train_accuracy = model.predict(X_train)
train_prediction = accuracy_score(X_train_accuracy,Y_train)
print(train_prediction)
```

Test Accuracy

```python
X_test_accuracy = model.predict(X_test)
test_prediction = accuracy_score(X_test_accuracy,Y_test)
print(test_prediction)
```
